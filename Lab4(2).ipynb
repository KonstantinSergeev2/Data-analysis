```python
# Пункт 1 - загрузка данных.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Dataf = pd.read_csv('sample_data/athlete_events.csv')
print("Размерность данных:", Dataf.shape)
print("\nПервые 10 строк данных:")
Dataf.head(10)
```

Функция shape показывает размерность датафрейма.

```python
# Пункт 2 - количество значений каждого из признаков.
print("Количество значений в каждом столбце:")
print(Dataf.count())
print("\nКоличество пропущенных значений:")
print(Dataf.isnull().sum())
```

Используем методы count и isnull().sum() для определения пропущенных данных.

```python
# Пункт 3 - статистический анализ.
statistic = Dataf[['Age', 'Height', 'Weight']].describe()
print("Статистическая информация:")
print(statistic)
```

Метод describe предоставляют необходимую статистику (среднее значение, стандартное отклонение, максимальное и минимальное значение, знаечние квартилей)

```python
# Пункт 4.1 - самый молодой участник  в 1992 году.
molodoy1992 = Dataf[Dataf['Year'] == 1992].nsmallest(1, 'Age')[['Name', 'Age', 'Event']]
print("Самый молодой участник ОИ 1992 года:")
display(molodoy1992)
```

nsmallest для получения самого маленького возраста.

```python
# Пункт 4.2 - уникальные виды спорта.
unique_sports = Dataf["Sport"].unique()
print(f"Всего уникальных видов спорта: {len(unique_sports)}")
print(unique_sports)
```

unique() даёт все уникальные виды спорта один раз.

```python
# Пункт 4.3 - средний рост теннисисток в 2000 году.
Average_tennis_2000 = Dataf[(Dataf["Year"] == 2000) & (Dataf["Sex"] == "F") & (Dataf["Sport"] == "Tennis")]
Average_height = Average_tennis_2000["Height"].mean()
print("Средний рост теннисисток в 2000 году (см):", Average_height)
```

Высчитываем средний рост теннисисток 2000 года.

```python
# Пункт 4.4 - количество золотых медалей, которые выиграл Китай в настольном теннисе в 2008.
China_gold = Dataf[(Dataf["Year"] == 2008) & (Dataf['NOC'] == 'CHN') & (Dataf['Sport'] == 'Table Tennis') & (Dataf['Medal'] == 'Gold')]
print(f"У Китая золотых медалей в настольном теннисе за 2008 год: {len(China_gold)}")
```

Просто фильтруем по году, стране, дисциплине, типу медали и всё, находим количество.

```python
# Пункт 4.5 - изменение количества видов спорта на летних ОИ 2004 года по сравнению с 1988.
Sport2004 = Dataf[(Dataf["Year"] == 2004) & (Dataf["Season"] == "Summer")]["Sport"].unique()
Sport1988 = Dataf[(Dataf["Year"] == 1988) & (Dataf["Season"] == "Summer")]["Sport"].unique()
print("Количество видов спорта в 2004 году:", len(Sport2004))
print("Количество видов спорта в 1988 году:", len(Sport1988))
print("Разница:", len(Sport2004) - len(Sport1988))
```

Считаем уникальные виды спорта, по заданным данным (год, сезон) и потом сравниваем.

```python
# Пункт 4.6 - гистограмма мужчин-керлингистов в 2014.
curling2014 = Dataf[(Dataf["Year"] == 2014) & (Dataf["Sport"] == "Curling") & (Dataf["Sex"] == "M")]
age = curling2014["Age"]
plt.figure(figsize=(10,5))
plt.title("Распределение возраста мужчин-керлингистов на ОИ 2014")
plt.xlabel("Возраст")
plt.ylabel("Количество спортсменов")
plt.hist(age)
plt.show()
```

hist - рисует график, title - название графика, xlabel - именование по иксу, ylabel - именование по игрику.

```python
# Пункт 4.7 - данные по медалям среди стран за 2006.
Winter2006 = Dataf[(Dataf["Year"] == 2006) & (Dataf["Season"] == "Winter")].copy()

Winter2006["Medal_Flag"] = Winter2006["Medal"].notna().astype(int)
Country_statistic = Winter2006.groupby("NOC").agg(medals_count = ("Medal_Flag", "sum"), sred_age = ("Age", "mean"))

Country_statistic = Country_statistic.sort_values("medals_count", ascending=False)
result2006 = Country_statistic[Country_statistic["medals_count"] > 0]
display(result2006.head(10))
```

groupby - сортировка по коду страны. Agg - функция для применения различных операций (суммирование, вычисление среднего и т.д). Ascending=False - Сортировка по убыванию.

```python
# Пункт 4.8 - таблица медалей за 2006.
pivot= pd.pivot_table(
    Winter2006,
    values='ID',
    index='NOC',
    columns='Medal',
    aggfunc='count',
    fill_value=0
)

pivot = pivot[['Gold', 'Silver', 'Bronze']]
print("Сводная таблица по медалям за зиму 2006:")
display(pivot.head(10))
```

Считаем медали по странам. Pivot_table - функция для создания сводных таблиц.

```python
Fotball2008 = Dataf[(Dataf["Year"] == 2008) & (Dataf["Sport"] == "Football")]

max_age = Fotball2008["Age"].max()

oldest_fotball = Fotball2008[Fotball2008["Age"] == max_age]

print("Возраст:", max_age)
display(oldest_fotball[["ID", "Name", "Age", "Team", "NOC", "Event"]])
```
