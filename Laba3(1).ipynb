```python
import numpy as np
# Пункт 1 - загрузка данных из файлов
np.set_printoptions(suppress=True)
generation = np.genfromtxt("sample_data/global-electricity-generation.csv", delimiter=",", dtype=float, skip_header=1, usecols=range(1,31))
consumption = np.genfromtxt("sample_data/global-electricity-consumption.csv", delimiter=",", dtype=float, skip_header=1, usecols=range(1,31))
countries = np.genfromtxt("sample_data/global-electricity-consumption.csv", delimiter=",", dtype=str, skip_header=1, usecols=0)

print("Страны:", countries, "\n", "Производство:", generation, "\n", "Потребление:", consumption, "\n")
```

Загружаем таблицы в три массива.

```python
# Пункт 2 - среднее производство и потребление электроэнергии (ежегодно) за 5 лет
Generation_last_5year = np.nanmean(generation[:, -5:], axis=1)
Consumption_last_5year = np.nanmean(consumption[:, -5:], axis=1)

print("Среднее производство за последние 5 лет:", Generation_last_5year, "\n", "Среднее потребление за последние 5 лет:", Consumption_last_5year, "\n")
```

[:, -5:], берём последние 5 лет. axis=1 для построчной обработки, np.nanmean считаем среднее арифметическое, игнорируем пропуски "nan".

```python
quantile5 = np.nanquantile(Consumption_last_5year, 0.05)

Countries_lowest = countries[Consumption_last_5year <= quantile5]
print("5% стран с наименьшшим потреблением за последние 5 лет:")
print(Countries_lowest)
```

np.nansum(consumption, axis=0) суммирует значения по столбцам.

```python
# Пункт 3.2 - максимальноое количество электроэнергии, которое произвела страна за один год.
Max_generation_in_year = np.nanmax(generation)
print("Максимальное количество произведённой электроэнергии за год (одной страной):", Max_generation_in_year)
```

np.nanmax(generation) игнорируем nan, выбираем максимальное число.

```python
# Пункт 3.3 - список стран, которые производят более 500 млрд. кВт*ч ежегодно в среднем за последние 5 лет.
Countries_More_500 = countries[Generation_last_5year > 500]

print("Страны с потреблением больше 500 млрд. кВт*ч:", Countries_More_500)
```

Воспользуемся массивом из второго пункта со средним производством электроэнергии и найдём интересующие нас страны.

```python
# Пункт 3.4 - 10% стран, которые потребляют больше всего электроэнергии ежегодно в среднем за последние 5 лет.
quantil = np.nanquantile(Consumption_last_5year, 0.9)
Countries_10_consumption = countries[Consumption_last_5year >= quantil]
print("10% стран с самым высоким потреблением электроэнергии за последние 5 лет", Countries_10_consumption)
```

np.nanquantile(Consumption_last_5year, 0.9) находим такие значения, где только 10% данных будут выше. Потом просто сравниваем и находим 10% стран с самым высоким потреблением.

```python
# Пункт 3.5 - Страны, которые увеличили прооизводство электроэнергии в 2021 по сравнению с 1992 в более, чем 10 раз.
generation_1992_year = generation[:, 0]
generation_2021_year = generation[:, -1]

condition = np.logical_and(generation_1992_year > 0, generation_2021_year > 10 * generation_1992_year)
Countries_generation_10x = countries[condition]

print("Страны, которые увеличили производство больше, чем в 10 раз", Countries_generation_10x)
```

np.logical_and выполняем логическую операцию, с условием, что производство в 2021 больше в 10 раз, чем в 1992

```python
# Пункт 3.6 - Страны, которые потратили в сумме за все годы больше 100 млрд. кВт*ч и при этом произвели меньше, чем потратили.
Sum_consumption = np.nansum(consumption, axis=1)
Sum_generation = np.nansum(generation, axis=1)

condition_generation = np.logical_and(Sum_consumption > 100, Sum_generation < Sum_consumption)
Countries_generation_less = countries[condition_generation]

print("Страны с суммарным потреблением > 100 млрд. кВт*ч и произвели меньше:", Countries_generation_less)
```

np.logical_and выполняем логическую операцию, с условиями, что суммарное потребление больше 100 и страна произвела меньше электроэнергии, чем потребила.

```python
# Пункт 3.7 - Какая страна потратила наибольшее количество электроэнергии в 2020 году?
Consumption_2020 = consumption[:, -2]

max_consumption_2020 = np.nanargmax(Consumption_2020)
country_max_consumption_2020 = countries[max_consumption_2020]
max_consumption_2020_value = Consumption_2020[max_consumption_2020]

print("Страна с наибольшим потреблением электроэнергии в 2020 году:", country_max_consumption_2020)
print("Потребление (млрд. кВт*ч):", max_consumption_2020_value)
```
