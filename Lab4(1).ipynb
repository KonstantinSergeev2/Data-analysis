```python
# Пункт 1 - Генерация случайного массива из 1000 значений НРСВ.
import numpy as np
import pandas as pd
M = 1.0
s = 1.0
N = 1000

array = pd.Series(np.random.normal(loc=M, scale=s, size=N))
print("Первые 10 значений (из 1000)")
print(array.head(10))
```

Генерация 1000 значений нормально распределённой случайной величины  с помощью np.random.normal. Полученный массив преобразован в pandas.Series.

```python
# Пункт 2 - Вычислить долю всех значений в диапазоне M-s; M+s
dolyaMS = ((array > (M - s)) & (array < (M + s))).mean()
print("Доля в (M-s, M+s):", dolyaMS, "или", dolyaMS * 100, "%")
```

Применяем булев массив с указанием диапазона, а для вычисления доли используем .mean из библиотеки Pandas.

```python
# Пункт 3 - какая доля всех значений находится в диапазоне M-3s; M+3s.
dolya3S = ((array > (M - 3*s)) & (array  < (M + 3*s))).mean()
print("Доля в (M-3s, M+3s):", dolya3S, "или", dolya3S * 100, "%")
```

По сути делаем то же самое, просто диапазон другой.

```python
# Пункт 4 - вычисление квадратного корня.
root = pd.Series(np.sqrt(array))
print("Первые 10 значений квадратного корня:")
print(root.head(10))
```

np.sqrt(array) берёт квадратный корень каждого элемента. Возникает предупреждения, так как из отрицательного значения, корень взять мы не можем, вместо этого значения пишется "NaN".

```python
# Пункт 5 - среднее арифметическое корней.
sred_root = root.mean()
print("Среднее арифметическое для корней:", sred_root)
```

Считаем среднее арифметическое корней.

```python
# Пункт 6 - создание DataFrame.
Dataf = pd.DataFrame({"number": array, "root": root})
print("Первые 6 строк из созданного DataFrame:")
print(Dataf.head(6))
```

Создаём табличку с помощью pd.DataFrame и выводим первые 6 строк, как по условию задания.

```python
# Пункт 7 - поиск значений, где корень 1.8-1.9.
result = Dataf.query("root >= 1.8 and root <= 1.9")
print("значения, где корень в диапазоне от 1.8 до 1.9:")
print(result)
```

Используем query, чтобы задать условия.
